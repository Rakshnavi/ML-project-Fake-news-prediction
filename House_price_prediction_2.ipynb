Import the dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
from sklearn.datasets import fetch_california_housing
```

Importing the house price dataset

```python
house_price_dataset = fetch_california_housing()
print(house_price_dataset)
```

```python
#loading the dataset to pandas dataframe
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
#print first five rows of dataset
house_price_dataframe.head()
```

```python
#add the target column to the dataframe
house_price_dataframe['price']=house_price_dataset.target
house_price_dataframe.head()
```

```python
#checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
#check for missing values
house_price_dataframe.isnull().sum()
```

```python
#statistical measure of the datasets
house_price_dataframe.describe()
```

Understanding the correlation between various features in the dataset

```python
#construct the heatmap to understand the correlation
correlation = house_price_dataframe.corr()

# Plot heatmap
plt.figure(figsize=(7, 7))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size': 5}, cmap='Blues')
plt.show()
```

Splitting the data and the target

```python
X=house_price_dataframe.drop(['price'],axis=1)
Y=house_price_dataframe['price']
print(X)
print(Y)
```

Splitting the data into training data and testing data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

Model training by
XGBoost Regressor

```python
#loading the model
model=XGBRegressor()
#training  the model with x_axis
model.fit(X_train,Y_train)
```

Evalution

Prediction on training data

```python
#accuracy for predicting on training data
training_data_prediction=model.predict(X_train)
print(training_data_prediction)
```

```python
# R squared error
score_1=metrics.r2_score(Y_train,training_data_prediction)
```

```python
# mean absolute error
score_2=metrics.mean_absolute_error(Y_train,training_data_prediction)
```

```python
print("R squared error for train data",score_1)
print("mean absolute error for train data",score_2)
```

Prediction on test data

```python
#accuracy for predicting on test data
test_data_prediction=model.predict(X_test)
print(test_data_prediction)
```

```python
# R squared error
scores_1=metrics.r2_score(Y_test,test_data_prediction)
```

```python
# mean absolute error
scores_2=metrics.mean_absolute_error(Y_test,test_data_prediction)
```

```python
print("R squared error for test data",scores_1)
print("mean absolute error for test data",scores_2)
```

Visualizing the actual prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual price Vs Predicted price")
plt.show()
```

Making a predictive system

```python
input_data = (7.2574, 52.0, 8.288136, 1.073446, 496.0, 2.802260, 37.85, -122.24)
# Convert input data into numpy array
input_data_as_numpy_array = np.array(input_data)
# Reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
# Make prediction
prediction = model.predict(input_data_reshaped)
print(prediction)
```
